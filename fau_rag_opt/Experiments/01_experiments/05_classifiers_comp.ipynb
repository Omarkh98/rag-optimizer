```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
summary_data = {
    "Model": [
        "TF-IDF + Logistic",
        "ST + Logistic",
        "ST + XGBoost",
        "ST + MLP"
    ],
    "Dense F1": [0.84, 0.81, 0.84, 0.85],
    "Hybrid F1": [0.75, 0.71, 0.69, 0.71],
    "Accuracy": [0.80, 0.77, 0.79, 0.80]
}
```

```python
df_summary = pd.DataFrame(summary_data)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.25
index = range(len(df_summary))

plt.bar([i - bar_width for i in index], df_summary["Dense F1"], bar_width, label='Dense F1', color='skyblue')
plt.bar(index, df_summary["Hybrid F1"], bar_width, label='Hybrid F1', color='salmon')
plt.bar([i + bar_width for i in index], df_summary["Accuracy"], bar_width, label='Accuracy', color='lightgreen')

# Labels and titles
plt.xlabel("Model")
plt.ylabel("Score")
plt.title("Classifier Performance Comparison")
plt.xticks(index, df_summary["Model"], rotation=15)
plt.ylim(0.6, 0.9)
plt.legend()
plt.tight_layout()

# Show plot and summary
print(df_summary), plt.show()
```
